Q23. Write a code that displays the sum of all the even numbers from the given list.
```
numbers = [12, 75, 150, 180, 145, 525, 50]
```

ANS. To display the sum of all the even numbers in a list, you can use the following code:



This code will iterate through the numbers in the list, and it will check if each number is even using the modulus operator (%). If the number is even, it will add it to the sum variable. After all of the numbers have been processed, the code will print the final value of sum.

In this case, the code will print the sum of all the even numbers in the list: 12 + 150 + 180 + 50 = 392.

````python
#CODE

numbers = [12, 75, 150, 180, 145, 525, 50]

# Initialize a variable to store the sum
sum = 0

# Iterate through the numbers in the list
for number in numbers:
    # Check if the number is even
    if number % 2 == 0:
        # Add the number to the sum
        sum += number

# Print the sum
print(sum)
````

Q25. Write a program to display only those numbers from a list that satisfy the following conditions

- The number must be divisible by five

- If the number is greater than 150, then skip it and move to the next number

- If the number is greater than 500, then stop the loop
```
numbers = [12, 75, 150, 180, 145, 525, 50]
```

To display only those numbers from a list that are divisible by five and meet the other specified conditions, you can use the following code:



This code will iterate through the numbers in the list, and it will use the continue and break statements to skip or stop the loop based on the specified conditions. If the number is divisible by five and meets the other conditions, it will be printed.

````python
numbers = [12, 75, 150, 180, 145, 525, 50]

# Iterate through the numbers in the list
for number in numbers:
    # Check if the number is divisible by five
    if number % 5 == 0:
        # Check if the number is greater than 500
        if number > 500:
            # Stop the loop
            break
        # Check if the number is greater than 150
        if number > 150:
            # Skip the number and move to the next one
            continue
        # Print the number
        print(number)
````
